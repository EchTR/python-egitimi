```python
import numpy as np
import pandas as pd
```

```python
m = np.random.randint(1, 100, size =(10,3))
df1 = pd.DataFrame(m, columns = ["var1","var2","var3"])
df1
```

```python
df2 = df1 + 100
df2
```

```python
pd.concat([df1, df2]) # bir index problemi oluyor çünkü ignore_index = False
```

```python
pd.concat([df1, df2], ignore_index = True) # ignore_index = True yaptığımız zaman, index problemi olmadan sonrasına yazıyor
```

```python
df1.columns = ["var1","var2","deg3"]
df1
```

```python
df2
```

```python
pd.concat([df1, df2]) # bazı sütunlar diğer dataframe'lerde olmayabiliyor. bu yüzden NaN değeri alıyoruz.
```

```python
pd.concat([df1, df2], join = "inner", ignore_index=True) # sadece ortak sütunlar.
```

```python
pd.concat([df1, df2], join_axes = [df1.columns])

# kullanımdan kaldırılmış :? 
# https://stackoverflow.com/questions/61362942/concat-got-an-unexpected-keyword-argument-join-axes
```
