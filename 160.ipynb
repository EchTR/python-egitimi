```python
import seaborn as sns
```

```python
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
sns.distplot(df.price, kde = False);
```

```python
sns.distplot(df.price, bins=100, kde=False); # bins değeri ne kadar fazla olursa grafik o kadar hassaslaşır
```

```python
sns.distplot(df.price); # kde kalkınca bir fonksiyon? geldi
```

```python
sns.distplot(df.price, hist = False);
```

```python
sns.kdeplot(df.price, shade = True);
```

```python
# 161. ders
(sns
    .FacetGrid(df,
        hue = "cut",
        height = 5,
        xlim = (0,10000))
    .map(sns.kdeplot, "price", shade=True)
 .add_legend()
);
```

```python
sns.catplot(x = "cut", y = "price", hue="color", kind="point", data = df);
```
