```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
df.describe().T
```

```python
df["sex"].value_counts()
```

```python
df["day"].value_counts()
```

```python
# boxplot, kursu izlerken aklıma gelen ve kompleks olduğunu düşündüğüm problemleri tek satırda çözdü
# https://tr.wikipedia.org/wiki/Kutu_grafiği
sns.boxplot(x = df["total_bill"]);
```

```python
sns.boxplot(y = df["total_bill"]);
```

```python
# hangi günler ne kadar kazanıyoruz?
sns.boxplot(x = "day", y = "total_bill", data = df);
```

```python
# bırakılan bahşiş kişi sayısına göre değişir mi?
sns.boxplot(x = "size", y = "total_bill", data = df);
```

```python
# üstteki grafikte çıkardığımız sonuca göre: evet, kişi sayısı arttıkça bırakılan bahşiş artıyor.
```

```python
sns.boxplot(x = "day", y = "total_bill", hue = "sex", data = df); 
# erkekler mi daha çok bahşiş bırakıyor kadınlar mı?
```

```python
# kadınlar çoğu zaman erkeklerden daha az bahşiş bırakma eğiliminde 
```

```python
# 165. bölüm
# https://yigitsener.medium.com/violin-plot-keman-grafi%C4%9Fi-ile-de%C4%9Fi%C5%9Fken-anlatma-sanat%C4%B1-8d8cebb72222
sns.catplot(y = "total_bill", kind = "violin", data = df);
```

```python
sns.catplot(x = "day", y = "total_bill", kind = "violin", data = df);
```

```python
sns.catplot(x = "day", y = "total_bill", kind = "violin", hue = "sex", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "time", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "day", style = "time", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "size", size = "size", data = df);
```
